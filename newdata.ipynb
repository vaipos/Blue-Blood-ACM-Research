Original Dataframe

```python
import boto3
import pandas as pd
import io

# Initialize S3 client
s3 = boto3.client('s3')

# Define bucket and key
bucket_name = 'blue-blood-data'
object_key = 'blue-blood-final.csv'

# Fetch the file from S3
response = s3.get_object(Bucket=bucket_name, Key=object_key)

# Read file content into a DataFrame
original_df = pd.read_csv(io.BytesIO(response['Body'].read()))

#dropping the columns
original_df = original_df.drop(columns = ['subject_id', 'pre_charttime', 'prescription_start', 'post_charttime'], axis = 1)


original_df


```
